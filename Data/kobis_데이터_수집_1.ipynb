```python
import pandas as pd
import requests
from concurrent.futures import ThreadPoolExecutor

# KOBIS 영화 상세 정보 조회 API URL
MOVIE_INFO_URL = "https://kobis.or.kr/kobisopenapi/webservice/rest/movie/searchMovieInfo.json"
API_KEY = "5425aa1e0e4236130596fdfb4dd00c47"

# 병렬 작업 함수
def fetch_movie_details(movie_code):
    if pd.notna(movie_code):  # 영화코드가 유효한 경우
        params = {
            "key": API_KEY,
            "movieCd": movie_code
        }
        try:
            response = requests.get(MOVIE_INFO_URL, params=params)
            if response.status_code == 200:
                data = response.json()
                movie_info = data.get("movieInfoResult", {}).get("movieInfo", {})

                # 배우 정보 (최대 5명까지 가져오기)
                actors = movie_info.get("actors", [])
                actor_names = ", ".join(actor.get("peopleNm", "N/A") for actor in actors[:5])

                # 상영시간
                show_time = movie_info.get("showTm", "N/A")

                # 관람등급
                audits = movie_info.get("audits", [])
                watch_grade = audits[0].get("watchGradeNm", "N/A") if audits else "N/A"

                return {"actors": actor_names, "show_time": show_time, "watch_grade": watch_grade}
        except Exception as e:
            print(f"에러 발생 (영화코드: {movie_code}): {e}")
    return {"actors": "N/A", "show_time": "N/A", "watch_grade": "N/A"}
```

```python
# 두 번째 행까지만 병렬 처리 실행
with ThreadPoolExecutor(max_workers=10) as executor:
    results = list(executor.map(fetch_movie_details, df2["영화코드"][:2]))  # 첫 두 행만 처리

# 결과 병합
df3.loc[:1, "배우"] = [result["actors"] for result in results]
df3.loc[:1, "상영시간"] = [result["show_time"] for result in results]
df3.loc[:1, "관람등급"] = [result["watch_grade"] for result in results]
```

```python
# 세 번째 행부터 병렬 처리 실행
with ThreadPoolExecutor(max_workers=10) as executor:
    results = list(executor.map(fetch_movie_details, df2["영화코드"][2:]))  # 세 번째 행부터 처리

# 결과 병합
df3.loc[2:, "배우"] = [result["actors"] for result in results]
df3.loc[2:, "상영시간"] = [result["show_time"] for result in results]
df3.loc[2:, "관람등급"] = [result["watch_grade"] for result in results]
```
